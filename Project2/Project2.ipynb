# 计算核矩阵并中心化标准化
## 核矩阵
只要一个对称函数对应的核矩阵半正定，它就能作为核函数。
核矩阵K总是半正定的：
![Image of Yaktocat](https://private.codecogs.com/gif.latex?K%20%3D%20%5Cbegin%7Bbmatrix%7D%20%5Ckappa%20%28x_%7B1%7D%2Cx_%7B1%7D%29%20%26%20.%20%26%20.%20%26%20.%20%26%20%5Ckappa%20%28x_%7B1%7D%2Cx_%7Bm%7D%29%20%5C%5C%20.%20%26%20.%20%26%20.%20%26%20.%20%26%20.%5C%5C%20.%20%26%20.%20%26%20.%20%26%20.%20%26%20.%5C%5C%20.%20%26%20.%20%26%20.%20%26%20.%20%26%20.%5C%5C%20%5Ckappa%20%28x_%7Bm%7D%2Cx_%7B1%7D%29%20%26%20.%20%26%20.%20%26%20.%20%26%20%5Ckappa%20%28x_%7Bm%7D%2Cx_%7Bm%7D%29%20%5Cend%7Bbmatrix%7D)
核方法往简单地说，就是将一个低维的线性不可分的数据映射到一个高维的空间、并期望映射后的数据在高维空间里是线性可分的。我们以异或数据集为例：在二维空间中、异或数据集是线性不可分的；但是通过将其映射到三维空间、我们可以非常简单地让其在三维空间中变得线性可分。

核抄矩阵是样本之间通过核函数影射之后得到的，每两个样本之间进行一次核函数影射，N个样本就构成N*N维的核矩阵。

使用书中5.1
$${\rm{K}}(x,y)=x^Ty$$

## 标准化
矩阵标准化的目的是，通过标准化处理，得到均值为0，标准差为1的服从标准正态分布的数据。（相对一维数据来说，也就是相对矩阵的每一列，数据的每一个维度）.

矩阵标准化使用如下公式完成：
![Image of Yaktocat](https://www.zhihu.com/equation?tex=x%5E%7B%27%7D%3D%5Cfrac%7Bx-%5Cmu+%7D%7B%5Csigma+%7D++)
μ表示均值，σ表示标准差，（u和σ都可以看成是行向量形式，它们的每个元素分表表示矩阵每一列的均值和方差）可以看出矩阵中心化是矩阵标准化的一步，将中心化的矩阵除以标准差得到标准化矩阵。

当原始数据不同维度上的特征的尺度（单位）不一致时，需要标准化步骤对数据进行预处理。
下图中以二维数据为例：左图表示的是原始数据；中间的是中心化后的数据，数据被移动大原点周围；右图将中心化后的数据除以标准差，得到为标准化的数据，可以看出每个维度上的尺度是一致的（红色线段的长度表示尺度）。
![Image of Yaktocat](https://pic3.zhimg.com/80/v2-d232db445418e9edf404767ef441789f_720w.jpg)


```python
import numpy as np
import pandas as pd

filepath = "C:\\Users\\Shinelon\\Desktop\\data-mining  project\\Project2\\iris.txt"
df = pd.read_csv(filepath)
# print(df)
df=df.iloc[:,[0,1,2,3]]  #除去最后一列
print(df)

df = df.values  #转换为数组
k=np.empty(shape=(149,149)) #生成空矩阵
for i in range(149):        #计算核矩阵
    for j in range(149):
        k[i][j]=np.dot(df[i],df[j])
print("核矩阵为：")
print(k)

mu = np.mean(k)   #计算均值
sigma = np.std(k) #计算标准差
cen1 = k - mu      #中心化矩阵
nor1 = (k-mu)/sigma#标准化矩阵
print("中心化矩阵为：")
print(cen1)
print("标准化矩阵为：")
print(nor1)
```

# 使用齐二次核将每个点变换到特征空间ϕ（x）
齐二次核矩阵$(\sqrt2*xy,x^2,y^2)$
可将四个属性扩展到十维
$(x_1{^2},x_2{^2},x_3{^2},x_4{^2},\sqrt2x_1x_2,\sqrt2x_2x_3, \sqrt2x_3x_4, \sqrt2x_1x_3, \sqrt2x_2x_4, \sqrt2x_1x_2)$

```python
m=np.empty(shape=((149,10)))
for i in range (149):
    for j in range(4):
        m[i][j] = df[i][j]*df[i][j] 
        m[i][4] = np.sqrt(2) * df[i][0] * df[i][1]
        m[i][5] = np.sqrt(2) * df[i][0] * df[i][2]
        m[i][6] = np.sqrt(2) * df[i][0] * df[i][3]
        m[i][7] = np.sqrt(2) * df[i][1] * df[i][2]
        m[i][8] = np.sqrt(2) * df[i][1] * df[i][3]
        m[i][9] = np.sqrt(2) * df[i][2] * df[i][3]
print("齐二次核矩阵：")
print(m)

mu=np.mean(m)    #均值
sigma=np.std(m)  #标准差
nor2=(m-mu)/sigma #标准化
cen2=m-mu
print("标准化齐二次核矩阵：")
print(nor2)
print("中心化齐二次核矩阵：")
print(cen2)

```
